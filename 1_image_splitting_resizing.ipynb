````python
df = []

for f in tqdm(files):
    cat = f.split('/')[-2]
    try: 
        img = Image.open(f)
        w, h = img.size

        df.append(
            dict(
                img_path=f,
                cat=cat,
                w=w,
                h=h,
                max_dim=max(w, h)
            )
        )
    except:
        continue
    
df = pd.DataFrame(df)
df
````

### Oversample the minorities

## Resize images

````python
export_dst = "/Users/cyc/Documents/python_files/SC201/project/resized"


for f in tqdm(files):
    try: 
        img = Image.open(f)
        w, h = img.size
        cat = f.split('/')[-2]
        os.makedirs(os.path.join(export_dst, cat), exist_ok=True)
        
        if w > h:
            w_dst = 1024
            h_dst = int(h / w * 1024)
        else:
            h_dst = 1024
            w_dst = int(w / h * 1024)
        
        
        img = img.resize((w_dst, h_dst), resample=Image.BILINEAR)
        
        img.save(os.path.join(export_dst, cat, os.path.basename(f)))
        
        
    except:
        continue
````
